The data set represents a **synthetic** one where **149,115** transactions in **coffee shops** of different locations were recorded accross diffent times of the day. The data set was conceptualized to not contain missing values.

For the purposes of this **Association Rule Analysis** experiment, only the `transaction_date`, `transaction_time`, `transaction_qty`, `store_id` and `product_type` will be kept.
`transaction_id` is not significant in the study as an ID represents a single item and not an actual transaction in the sense of a **basket**.

```python
sales_selected = sales.loc[:, ["transaction_date", "transaction_time", "transaction_qty", "store_id", "product_type"]]
sales_selected
```

`product_category` and `product_detail` could be used as well for the **Association Rule analysis**, but we preferred using `product_type` as it's closer to what day-to-day orders are called.

`transaction_date`, `transaction_time` and `store_id` will be used to spot individual basket items as those will have identical values for the previously stated features.

```python
product_types = set(sales["product_type"])
repr(product_types)
```

```python
print(f"Overall number of distinct product types: {len(product_types)}")
```
